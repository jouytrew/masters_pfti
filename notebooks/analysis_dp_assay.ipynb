```python
# Oritype and SamplingType have one value each, 'O' and 'Manual'
# This means these dimensions are not useful for analysis, and we really only have 17 dimensions

# we can drop those columns
dmlz_assay = dmlz_assay.drop(['Oritype', 'SamplingType'], axis=1)

# There are fewer sample ids and barcode numbers than expected, this will be looked at later on.
# This data suggests that there are 398 unique drillholes. This will need to be confirmed.
# Not sure what 'Tons_Sampling' is measuring, and why there are so many NaN values
# Need to check what units are used for 'SampleWeight'

# As can be seen, there is an inconsistency when it comes to variable names, 
# which could lead to confusion, so these should be renamed at some point to be more consistent.
```

```python
# there are 22378 unique sample ids, 3 fewer than expected.
# these 3 repeated IDs are at the corresponding indices.

u_samples = {}
for idx, row in dmlz_assay.iterrows():
    if row.SAMPLEID not in u_samples:
        u_samples[row.SAMPLEID] = [idx]
    else:
        u_samples[row.SAMPLEID].append(idx)
        
for k, vs in u_samples.items():
    if len(vs) > 1:
        print(f'{k}: {vs}')
```

```python
# all the repeated entries are adjacent. Entries with the same sample id have the same 'Tons_Sampling'
# They also share 'HOLEID' and 'DATESAMPLED', but the other values seem different, including 'BarcodeNo'

dmlz_assay.iloc[[16209, 16210, 18704, 18705, 18739, 18740]]
```

```python
# there are 22348 unique sample ids, 33 fewer than expected.
# these 33 repeated barcode numbers are at the listed indices below. 
# these indices are quite close together.

u_barcode = {}
for idx, row in dmlz_assay.iterrows():
    if row.BarcodeNo not in u_barcode:
        u_barcode[row.BarcodeNo] = [idx]
    else:
        u_barcode[row.BarcodeNo].append(idx)

# list of indices of interest
idxs = []
for k, vs in u_barcode.items():
    if len(vs) > 1:
        idxs.extend(vs)
        print(f'{k}: {vs}')
```

```python
# in this case, entries that share 'BarcodeNo's do not seem related, 
# not sharing many values that can be seen in this snapshot.

dmlz_assay.iloc[idxs[30:40]]
```
